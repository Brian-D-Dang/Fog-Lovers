```python
import pandas as pd
import matplotlib.pyplot as plt
import streamlit as st
import seaborn as sns
from glob import glob
from datetime import datetime

# Load data from multiple CSV files
data_files = glob('data/fogdata/*.csv')
df_list = [pd.read_csv(file) for file in data_files]
df = pd.concat(df_list)

# Convert timestamp to datetime and extract hour and day of week
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek

# Filter data where category is 1
df = df[df['category'] == 1]

# Create spider plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='hour', y='DoN', hue='day_of_week')
plt.title('Spider plot of category=1 frequency by hour of the day and day of week')
st.pyplot(plt)

# Create line plot of total fog frequency per day
df['date'] = df['timestamp'].dt.date
df_date = df.groupby('date').size()
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_date)
plt.title('Line plot of total fog frequency per day')
st.pyplot(plt)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_list[0]

df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
hourly_frequencies = df.groupby('hour').size()
data_files = glob('data/fogdata/*.csv')
df_list = [pd.read_csv(file) for file in data_files]

series = hourly_frequencies

num_vars = len(series.index)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

values = series.values.flatten().tolist()
values += values[:1] # repeat the first value to close the circle
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], series.index, color='grey', size=8)

ax.set_theta_offset(np.pi/2) # rotate the plot by 90 degrees
ax.set_theta_direction(-1) # make the plot go clockwise

ax.set_rlabel_position(30)

bars = ax.bar(angles[:-1], values[:-1], width=0.35, alpha=0.9, linewidth=2, edgecolor='none')

plt.show()
```
